### 2.1 Coeficiente de Variaci√≥n (CV%)

**Definici√≥n:**
```
CV% = (Desviaci√≥n Est√°ndar / Media) √ó 100
```

**Uso en este estudio:**  
Medir **heterogeneidad del AROPE** dentro de cada dimensi√≥n sociodemogr√°fica (edad, sexo, hogar, empleo).

**Interpretaci√≥n:**

| CV% | Nivel Heterogeneidad | Significado |
|-----|---------------------|-------------|
| **> 50%** | ALT√çSIMA | Dimensi√≥n **muy estratificante** (diferencias enormes entre categor√≠as) |
| **30-50%** | ALTA | Dimensi√≥n **importante** (diferencias significativas) |
| **15-30%** | MODERADA | Dimensi√≥n **relevante** (diferencias moderadas) |
| **< 15%** | BAJA | Dimensi√≥n **poco diferenciadora** (categor√≠as similares) |

**Ventajas del CV%:**
- ‚úÖ **Independiente de escala:** Permite comparar dimensiones con magnitudes diferentes
- ‚úÖ **Robusto:** No afectado por outliers como el rango
- ‚úÖ **Interpretable:** % tiene significado intuitivo

**Ejemplo de c√°lculo (Situaci√≥n Laboral):**

Datos AROPE 2023:
- Ocupados: 15.6%
- Parados: 62.1%
- Jubilados: ~14%
- Otros inactivos: ~18%

```python
import numpy as np

tasas = [15.6, 62.1, 14.0, 18.0]
media = np.mean(tasas)  # 27.425%
desv_std = np.std(tasas, ddof=1)  # 22.5pp
CV = (desv_std / media) * 100  # 82.0%
```

**Resultado:** CV = 82.0% ‚Üí ALT√çSIMA heterogeneidad  
**Conclusi√≥n:** Situaci√≥n laboral es **muy estratificante** (tener/no empleo determina fuertemente riesgo AROPE)

### 2.2 Impacto Redistributivo

**Definici√≥n:**
```
Impacto Redistributivo = Gini_antes_transferencias - Gini_despu√©s_transferencias
```

**Componentes:**
- **Gini antes de transferencias (con pensiones):** Desigualdad del mercado laboral + pensiones contributivas
- **Gini despu√©s de transferencias:** Desigualdad despu√©s de impuestos + transferencias sociales (IMV, prestaciones, etc.)

**Fuentes:**
- **Eurostat:** C√≥digos ilc_di12b (antes) y ilc_di12c (despu√©s)
- **Tablas:** `EUROSTAT_Impacto_Redistributivo_ES` y `EUROSTAT_Impacto_Redistributivo_UE27`

**Interpretaci√≥n:**
- **Mayor valor ‚Üí Estado redistribuye M√ÅS** (reduce m√°s la desigualdad del mercado)
- **Menor valor ‚Üí Estado redistribuye MENOS** (acepta desigualdad del mercado)

**Ejemplo de c√°lculo (Espa√±a 2024):**

```python
# Datos Espa√±a 2024 (aproximados)
gini_antes_ES = 37.2  # Desigualdad mercado
gini_despues_ES = 33.8  # Desigualdad final
impacto_ES = gini_antes_ES - gini_despues_ES  # 3.4pp

# Datos UE27 2024 (aproximados)
gini_antes_UE = 35.8
gini_despues_UE = 30.9
impacto_UE = gini_antes_UE - gini_despues_UE  # 4.9pp

# Diferencial
gap_redistributivo = impacto_ES - impacto_UE  # -1.5pp
```

**Resultado:** Espa√±a redistribuye **1.5pp MENOS** que UE27  
**Conclusi√≥n:** **Problema redistributivo doble:**
1. Mercado espa√±ol genera M√ÅS desigualdad (Gini antes: 37.2 vs 35.8)
2. Estado redistribuye MENOS (Impacto: 3.4pp vs 4.9pp)

**Causas identificadas:**
- Presi√≥n fiscal baja: 37% PIB (Espa√±a) vs 41% PIB (UE27)
- Gasto social bajo: 21% PIB (Espa√±a) vs 24% PIB (UE27)
- IMV tard√≠o (2020) con cobertura limitada (25%)
- Bajo gasto en familia: 1.3% PIB vs 2.3% PIB (UE27)

### 2.3 Convergencia Regional (Sigma y Beta)

#### 2.3.1 Convergencia Sigma (œÉ)

**Definici√≥n:**  
Mide si la **dispersi√≥n entre regiones disminuye** en el tiempo.

**F√≥rmula:**
```
œÉ_t = desviaci√≥n_est√°ndar(AROPE_ccaa) en a√±o t

Convergencia Sigma ‚Üí œÉ_t < œÉ_t-1 (dispersi√≥n disminuye)
Divergencia Sigma ‚Üí œÉ_t > œÉ_t-1 (dispersi√≥n aumenta)
```

**Interpretaci√≥n:**
| œÉ | Significado |
|---|-------------|
| Bajando | Las CCAA se **acercan** (convergencia territorial) |
| Subiendo | Las CCAA se **alejan** (divergencia territorial) |
| Estable | Brechas territoriales se **mantienen** |

**Ejemplo Espa√±a (2008-2023):**
```python
# Evoluci√≥n de la dispersi√≥n AROPE entre CCAA
sigma_2008 = 7.2  # Desviaci√≥n est√°ndar baja
sigma_2013 = 9.8  # Crisis ‚Üí aumenta divergencia
sigma_2023 = 8.9  # Mejora pero NO recupera nivel pre-crisis

# Interpretaci√≥n
divergencia_neta = sigma_2023 - sigma_2008  # +1.7pp
```

**Conclusi√≥n:** Espa√±a experimenta **divergencia territorial** desde 2008:
- Las CCAA est√°n M√ÅS alejadas que antes de la crisis
- Consolidaci√≥n de dos velocidades: Norte-Sur, Interior-Litoral

---

#### 2.3.2 Convergencia Beta (Œ≤)

**Definici√≥n:**  
Mide si las regiones **m√°s pobres crecen m√°s r√°pido** que las ricas.

**F√≥rmula:**
```
Regresi√≥n: Œî%AROPE_i = Œ± + Œ≤ ¬∑ AROPE_inicial_i + Œµ_i

Œ≤ < 0 ‚Üí Convergencia (pobres reducen m√°s)
Œ≤ > 0 ‚Üí Divergencia (pobres reducen menos)
Œ≤ = 0 ‚Üí Neutralidad (crecimiento independiente)
```

**Interpretaci√≥n:**
| Œ≤ | Significado |
|---|-------------|
| **Negativo** | CCAA pobres **recuperan terreno** (convergencia) |
| **Positivo** | CCAA pobres **pierden terreno** (divergencia) |
| **Cercano a 0** | **No hay relaci√≥n** entre nivel inicial y cambio |

**Ejemplo Espa√±a (2013-2023):**
```python
# Regresi√≥n simple
Œ≤_estimado = +0.12  # POSITIVO ‚Üí divergencia
p_value = 0.34      # NO significativo estad√≠sticamente

# Interpretaci√≥n
# No hay convergencia beta: las CCAA pobres NO reducen AROPE m√°s r√°pido
# El nivel inicial NO predice la mejora posterior
```

**Conclusi√≥n:**  
- **Divergencia sigma:** Las brechas se ampl√≠an (œÉ aumenta)
- **No convergencia beta:** Los pobres NO recuperan terreno
- **Resultado:** Espa√±a muestra **divergencia territorial sostenida**

### 2.4 AROPE (At Risk of Poverty or Social Exclusion)

**Definici√≥n:**  
Indicador multidimensional que identifica personas en riesgo de pobreza o exclusi√≥n social.

**F√≥rmula l√≥gica:**
```
AROPE = Pobreza_Monetaria OR Privaci√≥n_Material OR Baja_Intensidad_Laboral
```

Una persona est√° en **AROPE** si cumple **AL MENOS UNO** de estos tres criterios:

---

#### Componente 1: Pobreza Monetaria Relativa

**Umbral:** Ingresos < 60% mediana nacional

**C√°lculo:**
```python
# Ejemplo Espa√±a 2023
mediana_ingresos = 18_000 ‚Ç¨/a√±o equivalente
umbral_pobreza = 0.6 * mediana_ingresos  # 10,800 ‚Ç¨/a√±o

# Personas con ingresos < 10,800 ‚Ç¨ ‚Üí Pobreza monetaria
```

**Caracter√≠sticas:**
- **Relativo:** Depende de la mediana de cada pa√≠s/regi√≥n
- **Din√°mico:** Cambia con la distribuci√≥n de ingresos
- **Limitaci√≥n:** Ignora patrimonio, gastos, privaciones reales

---

#### Componente 2: Privaci√≥n Material Severa (2020+: Carencia Material y Social Severa - CMSS)

**Umbral:** No poder permitirse ‚â• 7 de 13 √≠tems b√°sicos

**√çtems (lista CMSS 2020+):**
1. Pagar alquiler, hipoteca o facturas
2. Mantener la vivienda caliente
3. Afrontar gastos imprevistos (‚âà 750‚Ç¨)
4. Comer carne/pescado cada dos d√≠as
5. Una semana de vacaciones fuera de casa al a√±o
6. Acceso a coche
7. Sustituir muebles viejos
8. Sustituir ropa vieja
9. Tener dos pares de zapatos en buen estado
10. Gastar peque√±a cantidad en s√≠ mismo semanalmente
11. Acceso a internet en casa
12. Sustituir electrodom√©sticos viejos
13. Tener actividades de ocio regulares

**Interpretaci√≥n:**
- Captura **privaciones reales** m√°s all√° del ingreso
- Refleja **pobreza estructural** y desconexi√≥n social
- M√°s estable que pobreza monetaria en crisis

---

#### Componente 3: Baja Intensidad Laboral en el Hogar (BITH)

**Umbral:** Intensidad laboral del hogar < 0.2 (menos del 20% de meses trabajados)

**C√°lculo:**
```python
# Ejemplo hogar con 2 adultos activos
# Adulto 1: trabaja 6 meses ‚Üí 6/12 = 0.5
# Adulto 2: trabaja 0 meses ‚Üí 0/12 = 0.0
# Intensidad hogar = (0.5 + 0.0) / 2 = 0.25 ‚Üí NO cumple BITH (>0.2)

# Hogar con 1 adulto activo
# Adulto 1: trabaja 2 meses ‚Üí 2/12 = 0.17 ‚Üí S√ç cumple BITH (<0.2)
```

**Caracter√≠sticas:**
- **Solo adultos 18-64 a√±os** (excluye jubilados, estudiantes)
- Captura **exclusi√≥n del mercado laboral**
- Relacionado con pobreza infantil (hogares sin empleo)

---

#### Agregaci√≥n AROPE

**Operador l√≥gico OR:**
```python
AROPE = (Pobreza_Monetaria == True) or \
        (Privacion_Material == True) or \
        (Baja_Intensidad_Laboral == True)
```

**Caracter√≠sticas:**
- **No aditivo:** 1 persona puede cumplir 2-3 condiciones
- **Multidimensional:** Captura diferentes formas de exclusi√≥n
- **Europeo:** Comparable entre pa√≠ses UE

### 2.5 Coeficiente de Gini y Ratio S80/S20

#### 2.5.1 Coeficiente de Gini

**Definici√≥n:**  
Medida de desigualdad en la distribuci√≥n de ingresos. Var√≠a de 0 (igualdad perfecta) a 100 (desigualdad m√°xima).

**F√≥rmula:**
```
Gini = (√Årea entre Lorenz y diagonal) / (√Årea total bajo diagonal)

Valores:
0-20: Igualdad muy alta (pa√≠ses n√≥rdicos)
20-30: Igualdad moderada (Europa Occidental)
30-40: Desigualdad moderada (Espa√±a, Italia, Portugal)
40-60: Desigualdad alta (LATAM, Sud√°frica)
```

**Interpretaci√≥n:**
| Gini | Escenario | Ejemplos |
|------|-----------|----------|
| < 25 | Igualdad alta | Eslovaquia (21), Finlandia (25) |
| 25-30 | Igualdad media-alta | Francia (28), Alemania (29) |
| 30-35 | Desigualdad media | **Espa√±a (33.8)**, Italia (34) |
| 35-40 | Desigualdad media-alta | Ruman√≠a (38), Bulgaria (39) |
| > 40 | Desigualdad alta | Sud√°frica (63), Brasil (52) |

**Ventajas:**
- **Resumen √∫nico:** Toda la distribuci√≥n en un solo n√∫mero
- **Comparable:** Entre pa√≠ses y periodos
- **Gr√°fico intuitivo:** Curva de Lorenz

**Limitaciones:**
- **Insensible a transferencias:** No distingue entre redistribuir de ricos a clase media vs a pobres
- **Oculta extremos:** Mismo Gini puede esconder diferentes estructuras (concentraci√≥n arriba vs abajo)

**Ejemplo Espa√±a 2024:**
```python
gini_ES = 33.8  # Desigualdad moderada
gini_UE27 = 30.9  # Espa√±a +2.9pp m√°s desigual que UE27

# Evoluci√≥n
gini_2008 = 31.9  # Pre-crisis
gini_2014 = 34.7  # Pico crisis
gini_2024 = 33.8  # Mejora pero NO recupera nivel pre-crisis
```

#### 2.5.2 Ratio S80/S20

**Definici√≥n:**  
Ratio entre el ingreso medio del **20% m√°s rico** (quintil superior) y el **20% m√°s pobre** (quintil inferior).

**F√≥rmula:**
```
S80/S20 = (Ingreso_medio_quintil_5) / (Ingreso_medio_quintil_1)

Ejemplo Espa√±a 2024:
Quintil 5 (20% m√°s rico): 40,000 ‚Ç¨/a√±o
Quintil 1 (20% m√°s pobre): 6,500 ‚Ç¨/a√±o
S80/S20 = 40,000 / 6,500 = 6.15
```

**Interpretaci√≥n:**
| S80/S20 | Significado |
|---------|-------------|
| < 4 | Desigualdad baja (los ricos ganan < 4x que los pobres) |
| 4-5 | Desigualdad media (Europa Occidental) |
| 5-7 | Desigualdad moderada-alta (Espa√±a, Italia, Portugal) |
| > 7 | Desigualdad alta (LATAM, Ruman√≠a, Bulgaria) |

**Ventajas:**
- **Intuitivo:** F√°cil de comunicar ("los ricos ganan 6x m√°s que los pobres")
- **Focalizado:** Captura brechas en extremos de la distribuci√≥n
- **Robusto:** Menos sensible a outliers extremos que Gini

**Limitaciones:**
- **Ignora el centro:** No ve qu√© pasa con el 60% intermedio
- **Solo extremos:** Cambios en quintiles 2-4 no se reflejan

**Ejemplo Espa√±a (2008-2024):**
```python
s80s20_2008 = 5.6  # Pre-crisis
s80s20_2014 = 6.8  # Pico crisis (+1.2 ‚Üí ricos ganan 1.2x m√°s que antes relativo a pobres)
s80s20_2024 = 6.2  # Mejora pero +0.6 por encima de 2008

# Comparaci√≥n europea 2024
s80s20_ES = 6.2
s80s20_UE27 = 4.7  # Espa√±a +1.5 puntos m√°s desigual
```
