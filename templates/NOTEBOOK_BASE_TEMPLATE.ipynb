## 3. Validación y Warnings Automáticos
Implementa funciones para validar resultados, detectar outliers y alertar sobre incoherencias.

```python
def validar_rangos(df, columna, min_val, max_val):
    outliers = df[(df[columna] < min_val) | (df[columna] > max_val)]
    if not outliers.empty:
        print(f'Warning: Valores fuera de rango en {columna}')
    return outliers
```

## 4. Comparación entre Fuentes o Versiones
Si tienes varias fuentes o versiones de datos, compara resultados y documenta discrepancias.

```python
def comparar_fuentes(df1, df2, columna):
    dif = df1[columna] - df2[columna]
    print(f'Diferencia media: {dif.mean()}')
    return dif
```
